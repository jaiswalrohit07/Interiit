```python


retracement_after_n_candle = []

def retrace(df, imbalance_time):    #need to add zone for better track of market
    ret_count = 0
    for i in imbalance_time:
        imb_price = df[df["Interval"]== i].close
#         imb_index = df[df["Interval"]== i].index.values[0]
        imb_index = df[df["Interval"]== i].index[0]

        
        if imb_index+1 < len(df):
#             if float(df.loc[imb_index+1].close.iloc[0]) > float(df.loc[imb_index].close.iloc[0]):
            if float(df.loc[imb_index+1].close) > float(df.loc[imb_index].close):

                print('up')
                
                for index, row in df.loc[imb_index+1:].iterrows():
#                     if float(df.loc[imb_index].close.iloc[0]) >= float(row.close):
                    if float(df.loc[imb_index].close) >= float(row.close):

                        ret_count += 1
                        retracement_after_n_candle.append(index - imb_index)
                        break
                        
                        
            else:
                print('down')
                for index, row in df.loc[imb_index+1:].iterrows():
#                     if float(df.loc[imb_index].close.iloc[0]) <= float(row.close):
                    if float(df.loc[imb_index].close) <= float(row.close):

                        ret_count += 1
                        retracement_after_n_candle.append(index - imb_index)
                        break
    return ret_count
            
                        

```
